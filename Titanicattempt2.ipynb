```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('titanic/train.csv')
df['Pclass'].max()
```

Cleaning the data

```python
df = df.drop(['PassengerId','Name','Cabin','Ticket'],axis=1,inplace=False)
#df.head()
```

```python
df_encoded = pd.get_dummies(df,columns=['Pclass','Embarked'],dummy_na=True)
df_encoded['Sex'] = df_encoded['Sex'].map({'male':0, 'female':1})
#df_encoded= df_encoded.astype(int)
#df_encoded.drop(['Sex','Pclass','Embarked'],axis=1,inplace=True) not neccessary
#clean = pd.concat([df,df_encoded],axis=1)
#df_encoded.info() 
```

Impute missing data


```python
##impute missing data

from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
df_encoded['Age'] = imputer.fit_transform(df_encoded[['Age']])

df_encoded.info()
```

SPlitting data to train it


```python
#Splitting model to train and scaling data

from sklearn.model_selection import train_test_split
#from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.layers import Dense 
from tensorflow.keras.models import Sequential
from tensorflow.keras.regularizers import l2

X = df_encoded.drop(['Survived',],axis=1,inplace=False).values
Y = df_encoded['Survived'].values

#scaling data 
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

#training with sci-kitlearn
#model = LogisticRegression()
X_train,X_test,Y_train,Y_test = train_test_split(X_scaled,Y,test_size=0.3,random_state=42)
#model.fit(X_train,Y_train)

## training with tensorflow
model = Sequential([
    Dense(8,activation='relu',input_shape=(np.shape(X_scaled)[1],)),
    #Dense(4,activation='relu'),
    Dense(1,activation='sigmoid', kernel_regularizer=l2(0.02))

])

model.compile(optimizer='sgd',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,Y_train,epochs=200,batch_size=32)

#prediction
Y_hat = model.predict(X_test)

Y_pred =[ 1 if i >0.5 else 0 for i in Y_hat]

print(accuracy_score(Y_test,Y_pred))
np.shape(X_test)
```
